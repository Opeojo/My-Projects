## Clustering

### Determine the Number of Clusters

#### Using Hierarchical Graph

```python
# choosing the UnitPrice and Spent for the analysis
X = df_num_scaled[['UnitPrice', 'Spent']]
```

```python
# linkage instance 
hier_clust = linkage(X, method='ward')
```

```python
plt.figure(figsize=(12,9))
plt.title('Hierarchical Clustering Graph')
plt.xlabel('Observations')
plt.ylabel('Distance')
dendrogram(hier_clust,
           truncate_mode='level',
           p = 5,
          show_leaf_counts=False,
          no_labels=True)
plt.show()
```

#### Using the Elbow Method

```python
def optimise_k_means(df, max_k):
    wcss = {}
    for k in range(1, max_k):
        kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
        kmeans.fit(X)
        wcss[k] = kmeans.inertia_
        print(f'The within cluster sum of squares for {k} clusters is {wcss[k]:.2f}')
        
    # Generate elbow plot
    fig = plt.subplots(figsize=(10,5))
    plt.plot(list(wcss.keys()), list(wcss.values()), 'o-', color='orange')
    plt.xlabel('Number of Clusters')
    plt.ylabel('Cluster Inertia')
    plt.grid(True)
    plt.title('Elbow Plot of KMeans')
    plt.show()
```

```python
n_clust = 11
optimise_k_means(X, n_clust)
```

#### Using the Silhouette Score

```python
def optimise_silhouette(df, max_k):
    s_score = {}
    for k in range(2, max_k):
        kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto').fit(X)
        classes = kmeans.predict(X)
        s_score[k] = silhouette_score(X, classes)
        print(f'The silhouette score for {k} clusters is {s_score[k]:.3f}')
        
    # visualise the silhouette graph
    fig = plt.subplots(figsize=(10,5))
    plt.plot(list(s_score.keys()), list(s_score.values()), 'o-', color='orange')
    plt.xlabel('Number of Clusters')
    plt.ylabel('Silhouette Values')
    plt.grid(True)
    plt.title('Silhouette Scores')
    plt.show()
```

```python
n_clust = 11
optimise_silhouette(X, n_clust)
```

### Applying KMeans Clustering

```python
# with the 3 clusters
kmeans = KMeans(n_clusters= 3, n_init='auto')
kmeans.fit(X)

# create the category column in the original datasets
col_name = 'Kmeans_3'
df_num_scaled[col_name] =  kmeans.labels_
df_num[col_name] = kmeans.labels_
```

```python
df_num_scaled.sample(7)
```

#### Visualising the Clusters

```python
# show clusters centres ( the 3 centroids)
centroids = kmeans.cluster_centers_
# print(kmeans.cluster_centers_, end='\n\n')
print(centroids)
```

```python
# function to display clusters
def visualise_clusters(df, label, *args):
    plt.scatter(df['UnitPrice'], df['Spent'], c=df[label], cmap='viridis')
    if args:
        plt.scatter(args[0][:,0], args[0][:,1], marker='*', color='black')
    plt.xlabel('Unit Price')
    plt.ylabel('Total Amount Spent')
    plt.title('Clusters of Customers')
    plt.legend()
    plt.show()
```

```python
# visualise the clusters by KMeans
visualise_clusters(df_num_scaled, col_name, centroids)
```

## Applying DBSCAN

```python
# apply DBSCAN and generate labels
dbscan = DBSCAN(eps=0.5, min_samples=20)
dbscan.fit(X)
labels = dbscan.labels_
print(np.unique(labels))
```

```python
#  add cluster labels to data
col_name = 'dbscan_labels'
df_num_scaled[col_name] = labels
df_num[col_name] = labels
```

```python
# visualise the clusters by DBSCAN
visualise_clusters(df_num, col_name)
```

### Applying Gaussian Mixture Model

```python
# initialise and fit the model
gmm = GaussianMixture(n_components=3, random_state=42)
gmm.fit(X)

# get the cluster labels and probabilities
gmm_labels = gmm.predict(X)
probs = gmm.predict_proba(X)
print(np.unique(gmm_labels))
```

```python
#  add cluster labels to data
col_name = 'Gmm_labels'
df_num_scaled[col_name] = gmm_labels
df_num[col_name] = gmm_labels
```

```python
# visualise the clusters detected by GMM
visualise_clusters(df_num, col_name)
```

## Evaluation

```python
# Calculate clustering metrics for KMeans
silhouette = silhouette_score(X, kmeans.labels_)
db_index = davies_bouldin_score(X, kmeans.labels_)
ch_index = calinski_harabasz_score(X, kmeans.labels_)

# Print the metric scores
print(f"Silhouette Score: {silhouette:.2f}")
print(f"Davies-Bouldin Index: {db_index:.2f}")
print(f"Calinski-Harabasz Index: {ch_index:.2f}")
```
